```python
ds_jobs_salaries = pd.read_csv("Data Science Jobs Salaries.csv")
ds_jobs_salaries.head()
```

```python
salaries = pd_dataset = pd.concat([ds_salaries, ds_jobs_salaries])
salaries['work_year'] =salaries['work_year'].fillna('').astype(str).str.rstrip('e')

salaries
```

```python
salaries.to_csv('salaries.csv', index=False)

moyenne_pays = salaries.groupby(['employee_residence', 'work_year']).agg('mean')
moyenne_pays = moyenne_pays.reset_index()
indexSA = moyenne_pays[moyenne_pays['employee_residence'] == 'SA'].index
indexUA = moyenne_pays[moyenne_pays['employee_residence'] == 'UA'].index
indexAE = moyenne_pays[moyenne_pays['employee_residence'] == 'AE'].index
moyenne_pays.drop(indexSA, inplace= True)
moyenne_pays.drop(indexUA, inplace= True)
moyenne_pays.drop(indexAE, inplace= True)
moyenne_pays['employee_residence'].unique()
code_pays = pd.read_csv('code_pays.csv', delimiter=";")

result_merged = pd.merge(moyenne_pays, code_pays, how='left', left_on= "employee_residence", right_on="Country_code")
result_merged.dropna(inplace=True)
moyenne_region = result_merged.groupby(['region_name', 'work_year']).agg('mean')


```
